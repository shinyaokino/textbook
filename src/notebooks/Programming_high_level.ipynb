```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 画像を読み込むためのライブラリをインポート
import PIL
```

```python
image = PIL.Image.open('holography_3750mg.JPG')
```

```python
plt.imshow(image)
```

```python
# クロッピング
interference = np.array(image)[500:2000, 1100:1800]
plt.imshow(interference)
```

```python
scale = np.array(image)[500:2000, 500:750]
plt.imshow(scale)
```

```python
# 中心の350〜450ピクセルを平均
inter_data = np.mean(interference[:,350:450,:], axis=1)
scale_data = np.mean(scale[:,150:250,:], axis=1)
```

```python
# RGB のうち Rについてplot
plt.figure(figsize=(10,8))
plt.plot(inter_data[:,0])
plt.plot(scale_data[:,1])
```
