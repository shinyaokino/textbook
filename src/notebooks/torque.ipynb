実験ノートを作る課題のコードです。

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
freq = [26.8,41.9,47.4,64.1,79.1,91.8,100.5,119.9,131.7,139.7,156.2,164.5,176,213.6,251.7,277.2,314,366.3,421.5,471.6,526.9,598.2,620.7]
force = [1.15,1.55,1.8,2.1,2.45,2.9,3.1,3.35,3.75,4,4.25,4.5,4.8,6.75,9.3,10.95,13.6,16.65,20.35,23.95,28.1,33.15,35.15]
```

```python
N = 13 # 傾向が変わるポイント
plt.plot(freq, force, 'o')
plt.plot(freq[:N], force[:N], 'o')
plt.xlabel('frequency [rpm]', size=14)
plt.ylabel('force [N]', size=14)
```

```python
# (0-N) の範囲、(N-)の範囲を別々にフィッティングする。
# (0-N) の範囲
a_low, b_low = np.polyfit(freq[:N], force[:N], 1)
a_hi,  b_hi  = np.polyfit(freq[N:], force[N:], 1)
```

```python
N = 13 # 傾向が変わるポイント
plt.plot(freq[:N], force[:N], 'bo')
plt.plot(freq[N:], force[N:], 'ro')
# フィッティング結果
plt.plot(np.linspace(0,300,10), [a_low*x + b_low for x in np.linspace(0, 300, 10)], 'b--')
plt.plot(np.linspace(150,700,10), [a_hi*x + b_hi for x in np.linspace(150, 700, 10)], 'r--')
plt.xlabel('rotation frequency [rpm]', size=14)
plt.ylabel('force [N]', size=14)
plt.savefig('../../src/figs/figs_python/torque.png')
```
